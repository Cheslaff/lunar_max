```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
slope = 2
bias = 1
```

```python
x = np.linspace(3.1, 26.8)
x = x.reshape(1, -1).T
y = x * slope + bias
```

```python
plt.scatter(x, y);
```

```python
from linear_models import LinearRegression
```

```python
model = LinearRegression(method="OLS")
model.fit(x, y)
```

```python
pred = model.predict(x)
```

```python
plt.scatter(x, y)
plt.plot(x, pred, c='red');
```

## OLS WORKS!

```python
model = LinearRegression(method="GD", learning_rate=0.0001)
model.fit(x, y)
```

```python
pred = model.predict(x)
plt.scatter(x, y)
plt.plot(x, pred, color="red");
```
